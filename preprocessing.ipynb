#### We have to group this information for each customer ID and take mean/sum for final

```python
data[['avg_loanamount','avg_totaldue']] = data[['avg_loanamount','avg_totaldue']].fillna(0)
```

#### Getting repayment days each customer took

```python
prevloan.head()
```

```python
test = prevloan['firstduedate']
```

```python
test.head()
```

```python
def createDate(strDate):
    strDate = strDate[:10]
    s = strDate.split('-')
    s = [int(y) for y in s]  
    s = date(s[0], s[1], s[2])
    return s
```

```python
prevloan['firstduedate'] = prevloan['firstduedate'].apply(createDate)
```

```python
prevloan['firstrepaiddate'] = prevloan['firstrepaiddate'].apply(createDate)
```

```python
prevloan['repaymentdays'] = prevloan['firstduedate'] - prevloan['firstrepaiddate']
```

```python
prevloan.head()
```

### Getting days in integer

```python
prevloan['repaymentdays'] = prevloan['repaymentdays'].astype('str').apply(lambda x: int(x.strip(' ')[:2]))
```

```python
prevloan.head()
```

```python
groupdata = prevloan[['customerid','repaymentdays']].groupby('customerid').sum()
groupdata = groupdata.reset_index()
groupdata.shape
```

```python
groupdata.head()
```

```python
data = pd.merge(data, groupdata, how='outer', on=['customerid'])
```

```python
data['repaymentdays'] = data['repaymentdays'].fillna(0)
```

#### Create balance train sample of size 600

```python
def getSample(size, df):
    size1 = int(round(size * 0.50))
    size2 = size - size1
    data_good = df[(df['good_bad_flag'] == 'Good')]
    data_bad = df[(df['good_bad_flag'] == 'Bad')]
    data_sample_good = data_good.sample(n=size1)
    data_sample_bad = data_bad.sample(n=size2)
    return pd.concat((data_sample_good, data_sample_bad))
```

```python
train_600 = getSample(600, train)
```

```python
train_600.to_csv('./dataset/train_600.csv', index=False)
```
